```python
import pandas as pd
import numpy as np

SESSIONPATH = 'Training Data/Session01/'
ARMDATA = 'armIMU.txt'
WRISTDATA = 'wristIMU.txt'
TIMEDATA = 'time.txt'


def load_data(filepath):
    
    arm = np.loadtxt(filepath + ARMDATA)
    wrist = np.loadtxt(filepath + WRISTDATA)
    time = np.loadtxt(filepath + TIMEDATA)
    #data = pd.DataFrame({'time':time, 'arm1':arm[:,1], 'wrist':wrist[:,1]})
    dataArm = pd.DataFrame(data=arm,
                       index=time,
                       columns=[1,2,3,4,5,6])
    dataWrist = pd.DataFrame(data=wrist,
                       index=time,
                       columns=[1,2,3,4,5,6])
    
    print(dataArm.shape)
    print(dataWrist.shape)
    print(dataWrist)
    
load_data(SESSIONPATH)    
```
